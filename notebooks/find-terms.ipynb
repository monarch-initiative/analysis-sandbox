```python
import requests
import csv
from json import JSONDecodeError

# Globals
BIOLINK_API = "https://api-dev.monarchinitiative.org/api"
NUM_TERMS = 4

# Input/output files
file_path = "/home/kshefchek/Downloads/MONDO_Coded_IC.tsv"
output_file = "/home/kshefchek/MONDO_Coded_IC-results.tsv"

outfile = open(output_file, 'w')

columns = ['score', 'description', 'semantic type', 'matched label', 'matched id', 'matched text']

outfile.write("#{}\n".format("\t".join(columns)))

biolink_search = BIOLINK_API + "/search/entity/autocomplete/"

parameters = {
    'rows': NUM_TERMS,
    'prefix': ['HP','MONDO','UBERON']
}

with open(file_path, 'r') as csv_file:
    reader = csv.reader(csv_file, delimiter='\t', quotechar='\"')
    next(reader, None)  # skip the headers
    for row in reader:
        if row[0].startswith('#'): continue
        (code, term, typ) = row[0:3]
              
        outfile.write("{0}\t{1}\t{2}\t".format(code, term, typ))
        
        if term == '':
            outfile.write("\t\t\t\n")
            continue
        
        url = biolink_search + term.replace("[", "").replace("]", "")
        biolink_request = requests.get(url, params=parameters)
        try:
            biolink_response = biolink_request.json()
        except JSONDecodeError:
            print(term)
            outfile.write("\t\t\t\n")
            continue
        
        for match in biolink_response['docs']:
            label = match['label'][0]
            match_id = match['id']
            matched_text = match['match']
            outfile.write("{}\t{}\t{}\t".format(label, match_id, matched_text))
        
        
        outfile.write("\n")

outfile.close()
```
